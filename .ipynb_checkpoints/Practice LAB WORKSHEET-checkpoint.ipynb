## Practical 1: Student Marks Analyzer

### Problem Definition
Store student data as a list of tuples. Remove duplicate marks, calculate average marks and store result in a dictionary.

### Sample Input
```python
students = [
    (1, "Amit", [78, 85, 90, 85]),
    (2, "Neha", [88, 92, 79, 88]),
    (3, "Ravi", [70, 75, 80, 75])
]
```

### Sample Output
```
{1: ('Amit', 84.33), 2: ('Neha', 86.33), 3: ('Ravi', 75.0)}

```

````python
students = [
    (1, "Amit", [78, 85, 90, 85]),
    (2, "Neha", [88, 92, 79, 88]),
    (3, "Ravi", [70, 75, 80, 75])
]

result={}
for stuid,name,marks in students:
    unique_marks = set(marks)
    avg = round(sum(unique_marks) / len(unique_marks),2)
    result[stuid] = (name,avg)

print(result)-+
````

## Practical 2: Inventory Management

### Problem Definition
Remove out-of-stock items and calculate total inventory value.

### Sample Input
```python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 100)
}
```

### Sample Output
```
{'pen': (10, 50), 'pencil': (5, 100)}
Total Value: 1000
```

````python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 100)
}

new_inventory = {}
total_value = 0

for items,(price,quantity) in inventory.items():
    if quantity > 0:
        new_inventory[items] = (price,quantity)
        total_value += price*quantity

print(new_inventory)
print('Total Value:',total_value)
````

## Practical 3: Email Domain Extractor

### Problem Definition
Extract email domains and count occurrences.

### Sample Input
```python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']
```

### Sample Output
```
{'gmail.com': 2, 'yahoo.com': 1}
```

````python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']

d_c = {}
li = []
for email in emails:
    domain = email.split('@')[1]
    li.append(domain) 
    d_c[domain] = li.count(domain)
    
print(d_c)
````

## Practical 4: Shopping Cart Billing

### Problem Definition
Calculate item-wise cost and total bill.

### Sample Input
```
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}
```

### Sample Output
```
{'apple': 100, 'banana': 30}
Total Bill: 130
```

````python
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}

i_c = {}
total_bill = 0

for item,qty in cart:
    cost = qty * prices[item]
    i_c[item] = cost
    total_bill += cost

print(i_c)
print("Total_bill:",total_bill)
````

## Practical 5: Character Position Index

### Problem Definition
Store positions of each character in a string.

### Sample Input
```python
s = "banana"
```

### Sample Output
```
{'b': [0], 'a': [1, 3, 5], 'n': [2, 4]}
```

````python


s = 'banana'
char_pos = {}

for i,c in enumerate(s):
    if c in char_pos:
        char_pos[c].append(i)
    else:
        char_pos[c] = [i]

print(char_pos)
````

## Practical 6: Survey Data Aggregation

### Problem Definition
Count responses city-wise and choice-wise.

### Sample Input
```python
responses = [
    ('Amit','Rajkot','Yes'),
    ('Neha','Rajkot','No'),
    ('Ravi','Ahmedabad','Yes')
]
```

### Sample Output
```
{'Rajkot': {'Yes': 1, 'No': 1}, 'Ahmedabad': {'Yes': 1}}
```

````python
responses = [
    ('Amit','Rajkot','Yes'),
    ('Neha','Rajkot','No'),
    ('Ravi','Ahmedabad','Yes')
]


res = {}
for name,city,choice in responses:
        if city not in res:
            res[city] = {}
        if choice in res[city]:
            res[city][choice] += 1
        else:
            res[city][choice] = 1

print(res)
````
