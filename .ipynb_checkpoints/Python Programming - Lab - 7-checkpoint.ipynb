### 01) WAP to find sum of tuple elements.
**Sample Input:**
```
T = (10, 20, 30, 40)

```
**Sample Output:**
```
Sum of tuple elements = 100
```

````python
T = tuple(map(int,input("Enter Elements:").split()))
sum=0
for i in range(len(T)):
    sum = sum + T[i]
print("Sum of Tuple Elements:",sum)
````

### 02) WAP to find Maximum and Minimum K elements in a given tuple.
**Sample Input:**
```
T = (7, 2, 9, 4, 1, 5)
K = 2

```
**Sample Output:**
```
Minimum 2 elements: (1, 2)
Maximum 2 elements: (9, 7)
```

````python
T = (7, 2, 9, 4, 1, 5)
K = 2

temp = list(T)

min_k = []
max_k = []

for _ in range(K):
    m = min(temp)
    min_k.append(m)
    temp.remove(m)

temp = list(T)


for _ in range(K):
    m = max(temp)
    max_k.append(m)
    temp.remove(m)

print("Minimum", K, "elements:", tuple(min_k))
print("Maximum", K, "elements:", tuple(max_k))
````

### 10) WAP to test if the given tuple is distinct or not.
**Sample Input:**
```
tuple1 = (1, 2, 3, 4)
tuple2 = (1, 2, 2, 3)
```
**Sample Output:**
```
(1, 2, 3, 4) is distinct? -> True
(1, 2, 2, 3) is distinct? -> False
```

````python
T1 = tuple(map(int,input("Enter Elements:").split()))

T2 = tuple(map(int,input("Enter Elements:").split()))
l1 = list(T1)
l2 = list(T2)
s1 = list(set(T1))
s2 = list(set(T2))

if l1 == s1:
    print(f"{T1} is Distinct->True")
else:
    print(f"{T2} is not Distinct ->False")

if l2 == s2:
    print(f"{T1} is Distinct->True")
else:
    print(f"{T2} is not Distinct ->False")


````

### 11) WAP to rotate the elements of a tuple to the right by `k` positions.

**Sample Input:**
```
T = (10, 20, 30, 40, 50)
k = 2
```
**Sample Output:**
```
(40, 50, 10, 20, 30)
```

````python
T = tuple(map(int,input("Enter Elements:").split()))
K = 2
a = list(T)

rotatedList = a[-K:] + a[:-K]
print(tuple(rotatedList))
````

### 12) WAP to merge two tuples of equal length alternately.

```
T1 = (1, 3, 5)
T2 = (2, 4, 6)
```
**Sample Output:**
```
(1, 2, 3, 4, 5, 6)
```

### 15) WAP to split a tuple into even-index and odd-index tuples.

**Sample Input:**
```
T = ('a', 'b', 'c', 'd', 'e')
```

**Sample Input:**
```
even = ('a', 'c', 'e')
odd = ('b', 'd')
```


````python
T = ('a', 'b', 'c', 'd', 'e')

even = []
odd = []

for i in range(len(T)):
    if i % 2 == 0:
        even.append(T[i])
    else:
        odd.append(T[i])

even_index = tuple(even)
odd_index = tuple(odd)

print("Even index tuple:", even_index)
print("Odd index tuple:", odd_index)
````

### 16) WAP to compute column-wise sum from a tuple of lists.

**Sample Input:**
```
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
```

**Sample Output:**
```
12 15 18

```

````python
t = ([1, 2, 3],
     [4, 5, 6],
     [7, 8, 9])

result = []

for col in zip(*t):
    result.append(sum(col))
    
print("Column-wise sum:", result)
````
