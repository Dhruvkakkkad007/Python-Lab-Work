### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
s = "Python3@Lab"
lw = 0
up=0
d=0
sp=0

for i in s:
    if i.islower():
        lw+=1
    elif i.isupper():
        up+=1
    elif i.isdigit():
        d+=1
    else:
        sp+=1
print("Lowercase :",lw)
print("Uppercase:" ,up)
print("Digits :",d)
print("Special Character :",sp)
```

### 2) Reverse Each Word
- Input: `Python is Easy` → `nohtyP si ysaE`
- Input: `Hello World` → `olleH dlroW`
- Input: `Learn Python Fast` → `nraeL nohtyP tsaF`

```python
s1 = input("Enter your String")
res=""
li = s1.split()
print(li)
for i in li:
    i = i[::-1]
    res+=i + " "

print(res)
```

### 3) Case Pattern Identifier
- `PYTHON` → Uppercase
- `python` → Lowercase
- `Python Programming` → Title Case

```python
s = input("Enter your String:")

if s.isupper():
        print(f"{s} Uppercase")
elif s.islower():
        print(f"{s} Lowercase")
elif s.istitle():
        print(f"{s} Title Case")
```

### 4) First and Last Occurrence Finder
- `programming`, `g` → First:3 Last:10
- `banana`, `a` → First:1 Last:5
- `hello`, `z` → Not found

```python
s = input("Enter your String:")
ch = input("Enter Character:") 
if s.find(ch)!=-1 and s.rfind(ch)!=-1:
    print(f"First Occurence:{s.find(ch)},Last Occurence:{s.rfind(ch)}")
else:
    print("Not Found")
```

### 7) Prefix and Suffix Removal
- `unhappy.txt` → `happy`
- `pretest.py` → `test`
- `unwanted.doc` → `wanted`

```python
s = input("Enter String:")
prefix = input("Enter Prefix:")
suffix = input("Enter Suffix:")

s = s.removeprefix(prefix).removesuffix(suffix)
print(f"Final String:{s}")
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
s = input("Enter String:")
ans = ""
for i in s:
        if i.islower():
            ans += i.upper()
        elif i.isupper():
            ans += i.lower()
print(ans)
    
```
