### 01) WAP to find their union, intersection, and difference of the given two sets.
**Sample Input:**
```
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

```
**Sample Output:**
```
Union: {1, 2, 3, 4, 5, 6}
Intersection: {3, 4}
Difference (A-B): {1, 2}

```

````python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

union_set = A.union(B)

intersection_set = A.intersection(B)

difference_set = A.difference(B)

print("Union:", union_set)
print("Intersection:", intersection_set)
print("Difference (A-B):", difference_set)


````

### 05) WAP to Merge two dictionaries. If a key exists in both dictionaries, add their values.

**Sample Input:**
```
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

```
**Sample Output:**
```
{1: 10, 2: 60, 3: 80, 4: 60}
```

````python
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

new_dict = Dict1.copy()

for key,value in Dict2.items():
    if key in new_dict:
        new_dict[key] += value
    else:
        new_dict[key] = value
print(new_dict)
````

### 06) WAP to Sort a dictionary in ascending order of its values.
Hint: in sorted(), use udf for key parameter

```
{'Math': 70, 'Physics': 85, 'Chemistry': 60}

```
**Sample Output:**
```
{'Chemistry': 60, 'Math': 70, 'Physics': 85}
```

````python
marks = {'Math': 70, 'Physics': 85, 'Chemistry': 60}

def getValue(item):
    return item[1]
    
d = sorted(marks.items(), key=getValue)

dict(d)
````

### 07) WAP to find the keys having maximum and minimum values in a dictionary.

**Sample Input:**
```
{'A': 50, 'B': 90, 'C': 30}
```
**Sample Output:**
```
Max Value Key: B
Min Value Key: C

```

````python
data = {'A': 50, 'B': 90, 'C': 30}

max_key = max(data, key=data.get)
min_key = min(data, key=data.get)

print("Max Value Key:", max_key)
print("Min Value Key:", min_key)
````

### 09) WAP to count the frequency of each character in a string using a dictionary.
**Sample Input:**
```
programming
```
**Sample Output:**
```
{'p': 1, 'r': 2, 'o': 1, 'g': 2, 'a': 1, 'm': 2, 'i': 1, 'n': 1}
```

````python
s = input("Enter string :")
freq = {}

for ch in s:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1
        
print(freq)
````

### 10) WAP to swap keys and values of a dictionary.
**Sample Input:**
```
{1: 'A', 2: 'B', 3: 'C'}
```
**Sample Output:**
```
{'A': 1, 'B': 2, 'C': 3}
```

````python
d = {1: 'A', 2: 'B', 3: 'C'}
d_new = {}
for key,value in d.items():
        d_new[value] = key

print(d_new)
````

### 11) WAP to find the student with the highest total marks, given student names and their marks.

**Sample Input:**
```
{'Amit': [70, 80, 90], 'Riya': [85, 75, 80], 'Neha': [60, 70, 65]}
```
**Sample Output:**
```
Topper: Amit
```

````python
students = {
    'Amit': [70, 80, 90],
    'Riya': [85, 75, 80],
    'Neha': [60, 70, 65]
}

max_total = 0
topper = ""

for name in students:
    total = sum(students[name])
    if total > max_total:
        max_total = total
        topper = name

print("Topper:", topper)
````

### 12) WAP to extract keys whose values are even numbers.

**Sample Input:**
```
{1: 11, 2: 20, 3: 15, 4: 40}
```
**Sample Output:**
```
[2, 4]
```

````python
d = {1: 11, 2: 20, 3: 15, 4: 40}
l1 = []
for key,value in d.items():
    if d[key]%2 == 0:
        l1.append(key)

print(l1)
````

### 13) WAP to Find common keys between two dictionaries.

**Sample Input:**
```
Dict1 = {1: 'A', 2: 'B', 3: 'C'}
Dict2 = {2: 'X', 3: 'Y', 4: 'Z'}

```
**Sample Output:**
```
[2, 3]
```

````python
Dict1 = {1: 'A', 2: 'B', 3: 'C'}
Dict2 = {2: 'X', 3: 'Y', 4: 'Z'}
common_keys = []
for key in Dict1:
    if key in Dict2:
        common_keys.append(key)
print(common_keys)
````

### 14) WAP to handle missing keys in dictionaries.

```
dict1 = {'a': 5, 'c': 8, 'e': 2}
key = 'c'
```

```
8

```
```
dict1 = {'a': 5, 'c': 8, 'e': 2}
```

### 15) WAP to create a dictionary using dictionary comprehension from two lists containing student names and marks. Include only those students in the dictionary whose marks are greater than or equal to 75.

**Sample Input:**
```
students = ['Amit', 'Riya', 'Neha', 'Karan']
marks = [78, 72, 85, 60]
```
**Sample Output:**
```
{'Amit': 78, 'Neha': 85}
```

````python
students = ['Amit', 'Riya', 'Neha', 'Karan']
marks = [78, 72, 85, 60]

result = {}

for i in range(len(students)):
    if marks[i] >= 75:
        result[students[i]] = marks[i]

print(result)
````
