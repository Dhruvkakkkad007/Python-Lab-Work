### 02) Create a class named Bank_Account with Account_No, User_Name, Email,Account_Type and Account_Balance data members. Also create a method GetAccountDetails() and DisplayAccountDetails(). Create main method to demonstrate the Bank_Account class.

```python
class Bank_Account:
    def __init__(self):
        self.Acc_No = "",
        self.User_Name = "",
        self.Email = "",
        self.Acc_Type = "",
        self.Acc_Balance = ""

    def GetAccDetails(self):
        self.Acc_No = input("Enter Acc No:"),
        self.User_Name = input("Enter User_Name:"),
        self.Email = input("Enter Email:"),
        self.Acc_Type = input("Enter Acc_Type:"),
        self.Acc_Balance = input("Enter Acc_Balance:")

    def Display(self):
        print(self.Acc_No),
        print(self.User_Name),
        print(self.Email),
        print(self.Acc_Type),
        print(self.Acc_Balance)

def main():
    b1 = Bank_Account()
    b1.GetAccDetails()
    b1.Display()

main()
```

### 03) WAP to create Circle class with area and perimeter function to find area and perimeter of circle.

```python
class Circle:
    def __init__(self):
        self.r = "",

    def area(self):
        self.r = float(input("Enter Radius"))
        a = 3.14*self.r*self.r
        print(a)

    def perimeter(self):
        self.r = float(input("Enter Perimeter"))
        p = 2*3.14*self.r
        print(p)


c1 = Circle()
c1.area()
c1.perimeter()        
```

### 04) Create a class for employees that includes attributes such as name, age, salary, and methods to update and display employee information.

```python
class Employee:
    def __init__(self,name,age,salary):
        self.name = name,
        self.age = age,
        self.salary = salary

    def update(self,new_name,new_age,new_salary):
        self.name = new_name,
        self.age = new_age,
        self.salary = new_salary
        print('Updated')

    def display(self):
        print("Name:",self.name),
        print("age:",self.age),
        print("salary:",self.salary)
        
e1 = Employee('Dhruv',18,25000)
e1.display()
e1.update('Heet',19,30000)
e1.display()
```

### 05) Create a bank account class with methods to deposit, withdraw, and check balance.

```python
class Bank_Acc:
    def __init__(self,name,balance):
        self.name = name,
        self.balance = balance

    def deposit(self,amt):
        self.balance+=amt
        print('Deposited:',self.balance)

    def withdraw(self,amt):
        if amt>self.balance:
            print('Insufficient Balance')
        else:
            self.balance-=amt
            print('Withdraw:',self.balance)
        
    def check_balance(self):
        print(f"Name:{self.name},Balance:{self.balance}")

b2 = Bank_Acc('Dhruv', 50000)
b2.check_balance()

b2.deposit(10000)
b2.check_balance()

b2.withdraw(5000)
b2.check_balance()
```
