### Practical 1: Count Lines, Words and Characters

**Problem Definition:**  
Write a Python program to read text from a file and count the total number of lines, words, and characters. The result should be written to an output file.


```python
with open("input.txt", "r") as file:
    text = file.read()   

lines = text.split("\n")
line_count = len(lines)

words = text.split()
word_count = len(words)

char_count = len(text)

with open("output.txt", "w") as file:
    file.write(f"Number of lines: {line_count}\n")
    file.write(f"Number of words: {word_count}\n")
    file.write(f"Number of characters: {char_count}\n")

print("Count completed.")
```


### Practical 2: Search a Word in a File

**Problem Definition:**  
Write a Python program to search a given word in a text file and count its occurrences. Store the result in an output file.


```python
search_word = input("Enter the word to search: ")

with open("que2.txt", "r") as file:
    text = file.read()

text = text.lower()
search_word = search_word.lower()

word_count = text.split().count(search_word)

with open("que2_op.txt", "w") as file:
    file.write(f"Word searched: {search_word}\n")
    
    file.write(f"Number of occurrences: {word_count}\n")

print("Search completed.")
```

### Practical 4: Maximum and Minimum using Tuple

**Problem Definition:**  
Write a Python program to read numbers from a file, store them in a tuple, and find the maximum and minimum values. Write the result to an output file.


```python
with open("que4.txt", "r") as file:
    data = file.read()

numbers = tuple(map(int, data.split()))

maximum = max(numbers)
minimum = min(numbers)

with open("que4_op.txt", "w") as file:
    file.write(f"Numbers: {numbers}\n")
    file.write(f"Maximum value: {maximum}\n")
    file.write(f"Minimum value: {minimum}\n")

print("Maximum and minimum values written ")
```


### Practical 5: Remove Duplicate Words using Set

**Problem Definition:**  
Write a Python program to read words from a file and remove duplicate words using a set. Store the unique words in an output file.


```python
with open("que5.txt", "r") as file:
    text = file.read()

words = text.split()

unique_words = set(words)

with open("que5_op.txt", "w") as file:
    for word in unique_words:
        file.write(word + " ")
```


### Practical 6: Word Frequency using Dictionary

**Problem Definition:**  
Write a Python program to count the frequency of each word in a text file using a dictionary and write the result to an output file.


```python
# Practical 6: Word Frequency using Dictionary

# Open input file
with open("input.txt", "r") as file:
    text = file.read()

# Convert text to lowercase
text = text.lower()

# Split text into words
words = text.split()

# Create empty dictionary
word_freq = {}

# Count frequency of each word
for word in words:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

# Write result to output file
with open("output.txt", "w") as file:
    for word, count in word_freq.items():
        file.write(f"{word} : {count}\n")

print("Word frequency written to output.txt")
```


### Practical 7: Student Result and Grade Calculation

**Problem Definition:**  
Write a Python program to read roll number, name, and marks of students from a file. Use a user-defined function to determine grade (A/B/C/Fail) and write the student result to an output file.

**Grade Criteria:**  
- Marks ≥ 75 : Grade A  
- Marks ≥ 60 : Grade B  
- Marks ≥ 40 : Grade C  
- Marks < 40 : Fail


```python
# Practical 7: Student Result and Grade Calculation

# User-defined function to determine grade
def calculate_grade(marks):
    if marks >= 75:
        return "A"
    elif marks >= 60:
        return "B"
    elif marks >= 40:
        return "C"
    else:
        return "Fail"

# Open input file
with open("input.txt", "r") as file:
    students = file.readlines()

# Open output file
with open("output.txt", "w") as out:
    out.write("RollNo  Name  Marks  Grade\n")

    for student in students:
        data = student.split()

        roll_no = data[0]
        name = data[1]
        marks = int(data[2])

        grade = calculate_grade(marks)

        out.write(f"{roll_no}  {name}  {marks}  {grade}\n")

print("Student results written to output.txt")
```
