### 04) WAP to find the longest word(s) in a file.

```python
fp = open("demo.txt","r")
data = fp.readlines()

maximum = data[0]

for i in data:
    if len(maximum) < len(i):
        maximum = i
    
print(maximum)
```

### 05) WAP to count the no. of lines, words and characters in a given text file.

```python
fp = open("demo.txt","r")
data = fp.readlines()
print(f"Lines: {len(data)}")

words = 0
char = 0
for i in data:
    words += len(i.split())
    char += len(i)


print("Words:",words)
print("Char:",char)
```

### 08) WAP to create an UDF named frequency to count occurances of the specific word in a given text file.

```python
def frequency(words):
    fp = open("demo.txt","r")
    c = 0
    for i in fp:
        c += i.split().count(words)
    print(c)

frequency("Hi")
```
