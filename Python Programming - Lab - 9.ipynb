### 02) Write a function that add first n numbers.

```python
n = int(input("Enter n:"))

def func(n):
    if n==0:
        return 0
    else:
        return n + func(n-1)

func(n)
```

### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
n = int(input("Enter n:"))
def isPrime(n):
    for i in range(2,n):
        if n%i==0:
            return 0       
    else:
        return 1

isPrime(n)
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
a = int(input("Enter a:"))
b = int(input("Enter b:"))

def Prime(a,b):
    li = list()
    for i in range(a,b+1):
        if i==1:
            continue
        if isPrime(i)==1:
           li.append(i)
        else:
            pass
    return li
    
Prime(a,b)
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
def fibonacci(n:int)->int:
    if n<=0:
        return 1
    if n==1:
        return 0
    if n==2:
        return 1
    return fibonacci(n-1)+fibonacci(n-2)

print(fibonacci(5))
```

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
s1 = "AbcDEfgh" 
def upperLower(s1):
    lower=0
    upper=0
    for i in s1:
        if i.islower():
            lower+=1
        else:
            upper+=1
    print(f"no_upper:{upper},no_lower:{lower} ")
upperLower(s1)
```

### 14) For the given list of names of students, extract the names having more that 7 characters. Use filter().

```python
students = ["Ajay", "Rahul", "Pratiksha", "Samantha", "Jay", "Pooja"]

result = list(filter(lambda students: len(students) > 7, students))

print(result)
```
