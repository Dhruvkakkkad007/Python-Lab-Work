### 01) WAP to find sum of tuple elements.

### 03) WAP to find tuples which have all elements divisible by K from a list of tuples.
**Sample Input:**
```
tuple_list = [(2, 4, 6), (3, 6, 9), (4, 8, 12), (5, 10, 15)]
K = 2

```
**Sample Output:**
```
Tuples divisible by 2 : [(2, 4, 6), (4, 8, 12)]
```

````python
lst = [(2, 4, 6), (3, 6, 9), (2, 3, 4), (8, 10, 12)]
K = 2

result = []

for t in lst:
    count = 0
    
    for num in t:
        if num % K == 0:
            count += 1
    
    if count == len(t):
        result.append(t)

print("Tuples divisible by 2:", result)
````

### 05) WAP to find tuples with all positive elements from the given list of tuples.

### 14) WAP to create (index, value) pairs for prime numbers.

**Sample Input:**
```
T = (4, 7, 9, 11, 15)
```
**Sample Output:**
```
((1, 7), (3, 11))
```

````python
T = (4, 7, 9, 11, 15)

result = []

for i in range(len(T)):
    num = T[i]
    
    if num > 1:
        for j in range(2, num):
            if num % j == 0:
                break
        else:
            result.append((i, num))

print(tuple(result))
````

### 15) WAP to split a tuple into even-index and odd-index tuples.

```
T = ('a', 'b', 'c', 'd', 'e')
```

**Sample Input:**
```
even = ('a', 'c', 'e')
odd = ('b', 'd')
```


````python
T = ('a', 'b', 'c', 'd', 'e')

even = []
odd = []

for i in range(len(T)):
    if i % 2 == 0:
        even.append(T[i])
    else:
        odd.append(T[i])

even_index = tuple(even)
odd_index = tuple(odd)

print("Even index tuple:", even_index)
print("Odd index tuple:", odd_index)
````

### 16) WAP to compute column-wise sum from a tuple of lists.

**Sample Input:**
```
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
```

**Sample Output:**
```
12 15 18

```

````python
t = ([1, 2, 3],
     [4, 5, 6],
     [7, 8, 9])

result = []

for col in zip(*t):
    result.append(sum(col))
    
print("Column-wise sum:", result)
````
