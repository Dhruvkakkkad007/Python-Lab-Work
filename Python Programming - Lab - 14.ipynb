### 04) WAP to plot a line chart showing temperature variation during a week. Mark the lowest and highest temperatue.

```python
Days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
Tempreture  = [20, 32, 31, 33, 34, 43, 36]
```

```python
plt.plot(Days,Tempreture,marker='o')
max_t = max(Tempreture)
min_t = min(Tempreture)
plt.scatter(Days[Tempreture.index(max_t)],max_t)
plt.scatter(Days[Tempreture.index(min_t)],min_t)
plt.text(Days[Tempreture.index(max_t)], max_t, "High")
plt.text(Days[Tempreture.index(min_t)], min_t, "Low")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Temperature Variation ")
plt.show()
```

### 05) WAP to create a detailed pie chart representing the monthly expenditure distribution of a family. 
### The chart should display labels, percentage values, title, different colors for each slice, and explode one slice to highlight the highest expense.

```python
Categories = ["Rent", "Food", "Education", "Transport", "Entertainment", "Savings"]
Amount = [15000, 8000, 6000, 3000, 2000, 5000]
```

```python
clr = ['pink','r','y','g','m']
plt.pie(Amount,labels = Categories ,radius = 1,autopct = '%1.2f%%',colors=clr,explode=[0.5,0.8,0.1,0.3,0.5,0.1],shadow=True)
plt.title('Expense')
plt.show()
```
