### 07) WAP to print sum of series 1 – 2 + 3 – 4 + 5 – 6 + 7 ... n.

```python
n = int(input("Enter a Number:"))
i=1
sum=0
while i<=n:
    if i%2==0:
        sum-=i
    else:
        sum+=i
    i+=1
print(f"Sum of Series:{sum}")
```

### 13) WAP to print sum of digits of given number.

### 14) WAP to check whether the given number is Palindrome or not.

```python
import math
n = int(input("Enter a Number:"))
sum=0
temp=n
while n>0:
    r=n%10
    sum=sum*10+r
    n=math.floor(n/10)

if temp==sum:
    print(f"{temp} is a Palindrome Number")
else:
    print(f"{temp} is not a Palindrome Number")
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
import math
n = int(input("Enter a Number:"))
temp=n
sum=0
while n>0:
    r=n%10
    sum=sum+(r*r*r)
    n=math.floor(n/10)

if temp==sum:
    print(f"{temp} is a Armstrong Number")
else:
    print(f"{temp} is not a Armstrong Number")
```
