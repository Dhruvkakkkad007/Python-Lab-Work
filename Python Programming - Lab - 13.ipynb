### 02) WAP to pick a random character from a given String.

```python
import random

s = input("Enter a String: ")

index = random.randint(0, len(s) - 1)

print("Random character:", s[index])
```

### 04) WAP to roll a dice in such a way that every time you get the same number.

```python
import random

n = int(input("Enter how many time you want to roll a dice :"))

for _ in range(n):
    random.seed(60)
    print("Dice roll:", random.randint(1,6))
```

### 05) WAP to generate 3 random integers between 100 and 999 which is divisible by 5.

```python
import random
n = int(input("Enter how many numbers you want between 100 and 999 which is divisible by 5 :"))
for _ in range(n):
    nums = random.randrange(100, 999, 5)
    print(nums , end = " ")
```

### 06) WAP to generate 100 random lottery tickets and pick two lucky tickets from it and announce them as Winner and Runner up respectively.

```python
import random

tickets = list(range(1, 101))

# winner , runner_up = random.choices(tickets, k=2)
winner , runner_up = random.sample(tickets, 2)
print("Winner Ticket:", winner)
print("Runner Up Ticket:", runner_up)
```

### 08) Subtract a week (7 days) from a given date in Python.

```python
import datetime as dt

before_date = dt.datetime.now()
after_date = before_date+dt.timedelta(days = -7)
print("todays date :",before_date)
print("after Subtracting 7 days or a week :",after_date)
```
