# Exception Handling

### 01) WAP to handle following exceptions:
1. ZeroDivisionError
2. ValueError
3. TypeError
#### Note: handle them using separate except blocks and also using single except block too.

```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    
    result = num1 / num2
    print("Result:", result)

except ZeroDivisionError as e:
    print("Error: Cannot divide by zero." , e)

except ValueError as e:
    print("Error: Invalid input! Please enter integers only." , e)

try : 
    num1 = int(input("Enter first number: "))
    num2 = 'aa'
    result = num1 / num2
    print("Result:", result)
    
except TypeError as e:
    print("Error: Type mismatch occurred." , e)
```

### 05) WAP to demonstrate else  and finally block.

```python
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    
    result = num1 / num2

except ZeroDivisionError as err:
    print("Error:" , err)

except ValueError:
    print("Error: " , err)

else:
    print("Division successful.")
    print("Result:", result)

finally:
    print("This block always executes.")
```

### 06) WAP to create an udf divide(a,b) that handles ZeroDivisionError.

```python
def divide(a,b):
    try:
        result = a / b
    except ZeroDivisionError as err:
        print("Error :" , err)
    else :
        return result

try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
except Exception as err:
    print("Error :" , err)
else :
    print("Division : ", divide(num1 , num2))
finally :
    print("Program Completed")
```

### 07) WAP to accept item prices and calculate total. Raise ValueError if price is negative, otherwise print the total bill amount.

```python
try :
    total_items = int(input("Enter Total Items :"))
    prices = []
    for i in range (total_items):
        temp = int(input("Enter price of Item :"))
        if temp < 0 :
            raise ValueError("Price can't be Negative")
        else :
            prices.append(temp)
except ValueError as err :
    print("Error :" , err)
except Exception as err :
    print("Error :" , err) ## for any other errors if they occur
else :
    print("total bill : " , sum(prices))
```

### 08) Create a short program that prompts the user for a list of grades separated by commas. 
#### Split the string into individual grades and use a list comprehension to convert each string to an integer. 
#### You should use a try statement to inform the user when the values they entered cannot be converted.

```python
try:
    marks = list(map(int , input("Enter grades separated by commas").split(",")))
    
except ValueError as err :
    print("Error : values entered cannot be converted to integer")
except Exception as err:
    print("Error :" , err)
else :
    print(marks)
finally:
    print("Program Completed")
```

### 09) Accept 5 subject marks (0-100). Use assert for validation.

#### Sample Input (Error Case)
#####  Enter marks: 110 

#### Sample Output (Error Case)
##### AssertionError: Marks should be between 0 and 100.

#### Sample Input (Normal Case)
##### Valid marks entered

#### Sample Output (Normal Case)
##### Average Marks displayed

```python
try :
    marks = []
    for i in range(5):
        temp = int(input("Enter mark of Subject :"))
        assert 0<=temp and temp<=100, "Marks should be between 0 and 100."
        marks.append(temp)
except AssertionError as err :
    print("Assertion Error :" ,err)
except Exception as err:
    print("Error :" ,err)
else:
    print("Average Marks :" , sum(marks)/5)
```

### 10) WAP that gets password from user. Password must have at least 8 characters and one digit. 
### Raise WeakPasswordError if invalid, print the messsage "Correct" otherwise.

```python
class WeakPasswordError(Exception):
    pass

try :
    password = input("Enter Password :")
    if len(password)<8:
        raise WeakPasswordError("Password must have at least 8 characters.")
    if not any(i.isdigit() for i in password):
        raise WeakPasswordError("Password must have at least 1 Digit.")
except WeakPasswordError as err:
    print("WeakPasswordError :" ,err)
except Exception as err:
    print("Error : " ,err)
else :
    print("Correct.")
```

### 11) WAP to raise your custom Exception named NegativeNumberError with the error message : "Cannot calculate the square root of a negative number" :
#### if the given number is negative.
#### otherwise print the square root of the given number.

```python
import math
class NegativeNumberError(Exception):
    pass
try:
    num=int(input("Enter number for Square Root:"))
    if num<0:
        raise NegativeNumberError("Cannot calculate the square root of a negative number")
except NegativeNumberError as err:
    print("NegativeNumberError :" , err)
except Exception as err:
    print("Error : " ,err)
else:
    print(math.sqrt(num))
```
