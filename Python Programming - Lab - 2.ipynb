### 11) WAP to accept two integers and display sum, difference, and product using f-strings.

### 12) WAP to accept date in dd mm yyyy format and display in multiple formats.
#### Input : 01 12 2025
#### Output :
##### 01/12/2025
##### 2025-12-01

```python
date = input("Enter date:")
date1 = date.replace(' ','/')
print(date1)
first,second,third = date.split(' ')
date = third+'-'+second+'-'+first
print(date)
```

### 14) WAP to convert degree into Fahrenheit and vice versa.

```python
deg = int(input("Enter degree:"))
f = (deg*9/5)+32
print("degree to Fahrenheit = ",f)

f = int(input("Enter degree:"))
deg = (f-32)*5/9
print("Fahrenheit to degree  = ",deg)
```
