### 02) Create a class named Bank_Account with Account_No, User_Name, Email,Account_Type and Account_Balance data members. Also create a method GetAccountDetails() and DisplayAccountDetails(). Create main method to demonstrate the Bank_Account class.

```python
class Bank_Account:
    def __init__(self, acc_no, name, email, acc_type, balance):
        self.Account_No = acc_no
        self.User_Name = name
        self.Email = email
        self.Account_Type = acc_type
        self.Account_Balance = balance

    def GetAccountDetails(self):
        return {
            "Account No": self.Account_No,
            "User Name": self.User_Name,
            "Email": self.Email,
            "Account Type": self.Account_Type,
            "Balance": self.Account_Balance
        }

    def DisplayAccountDetails(self):
        details = self.GetAccountDetails()
        output = ""
        for key, value in details.items():
            output += f"{key}: {value}\n"
        
        return output


acc = Bank_Account(101, "Dhruv", "dhruv@gmail.com", "Saving", 5000)
print(acc.DisplayAccountDetails())
```

### 03) WAP to create Circle class with area and perimeter function to find area and perimeter of circle.

```python
import math

class Circle:
    def __init__(self, r):
        self.r = r

    def area(self):
        return math.pi * self.r ** 2

    def perimeter(self):
        return 2 * math.pi * self.r
c1 = Circle(5)
print(c1.area())
print(c1.perimeter())
```

### 05) Create a bank account class with methods to deposit, withdraw, and check balance.

```python
class Bank:
    def __init__(self, balance=0):
        self.balance = balance

    def deposit(self, amt):
        if amt > 0:
            self.balance += amt

    def withdraw(self, amt):
        if amt <= self.balance:
            self.balance -= amt

    def check(self):
        return self.balance
b1= Bank(5000)
print(b1.check())
b1.deposit(5000)
print(b1.check())
b1.withdraw(5000)
print(b1.check())
b1.withdraw(5000)
print(b1.check())
b1.withdraw(5000)
print(b1.check())
```

### 06) Create a class for managing inventory that includes attributes such as item name, price, quantity, and methods to add, remove, and update items.

```python
class Inventory:
    def __init__(self, name, price, quantity):
        self.items = {}
        self.items[name] = [price, quantity]

    def add_item(self, name, price, quantity):
        self.items[name] = [price, quantity]
        return "Item added"

    def remove_item(self, name):
        if name in self.items:
            self.items[name][1] = 0
            return "Item removed"
        return "Item not found"

    def update_item(self, name, price, quantity):
        if name in self.items:
            self.items[name] = [price, quantity]
            return "Item updated"
        return "Item not found"

    def show_items(self):
        active_items = []
        for name in self.items:
            if self.items[name][1] > 0:
                active_items.append((name, self.items[name][0], self.items[name][1]))
        return active_items


inv = Inventory("Pen", 10, 50)
print(inv.add_item("Notebook", 40, 20))
print(inv.show_items())
print(inv.update_item("Pen", 12, 60))
print(inv.remove_item("Notebook"))
print("After update and remove:")
print(inv.show_items())
```

### 09) Define Time class with hour and minute as data member. Also define addition method to add two time objects.

```python
class Time:
    def __init__(self, hour, minute):
        self.hour = hour
        self.minute = minute

    def add(self, other):
        total_minutes = self.minute + other.minute
        extra_hour = total_minutes // 60
        minute = total_minutes % 60

        hour = self.hour + other.hour + extra_hour
        return Time(hour, minute)

    def show(self):
        print(f"{self.hour:02d}:{self.minute:02d}")


t1 = Time(2, 45)
t2 = Time(3, 35)
t3 = t1.add(t2)
print("Added Time:")
t3.show()
```
